```python
#Group by product name
df.groupby('product_name')
```

```python
#aggregating order number , mean
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#alternative for mean
df.groupby('department_id')['order_number'].mean()
```

```python
#multiple aggregates
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

#### Aggregate with transform()

```python
#Transform function
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check output
df.head(15)
```

#### loc() function

```python
#If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”

df.loc[df['max_order']>40, 'loyalty_flag'] = 'loyal_customer'
```

```python
#If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”

df.loc[(df_ords_prods_merge['max_order']>10) &(df['max_order']<=40) , 'loyalty_flag'] = 'regular_customer'
```

```python
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

df.loc[df_ords_prods_merge['max_order']<=10, 'loyalty_flag'] = 'new_customer'
```
